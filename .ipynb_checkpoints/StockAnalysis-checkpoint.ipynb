```python
scaled = plw["close"].values
train_size = int(len(scaled) * 0.9)
test_size = len(scaled)- train_size
train,test = scaled[0:train_size], scaled[train_size:len(scaled)]
```

```python
def univariate_data(dataset, history_size, target_size):
    data = []
    labels = []

    start_index = history_size
    #warning, for end iteration should be reconsidered referring to 
    
    for i in range(start_index, len(dataset)):
        indices = range(i-history_size, i)
        # Reshape data from (history_size,) to (history_size, 1)
        data.append(np.reshape(dataset[indices], (history_size, 1)))
        labels.append(dataset[i+target_size])
    return np.array(data), np.array(labels)


def transform(dataset, passed_tuple=None):
    if passed_tuple:
        mean, std = passed_tuple
        return (dataset - mean) / std, mean, std
    else:
        dataset_mean = dataset.mean()
        dataset_std = dataset.std()
    return (dataset-dataset_mean)/dataset_std, dataset_mean, dataset_std


def revert_transform(dataset, passed_tuple):
    mean,std = passed_tuple
    return dataset*std+mean
        
```

```python
train_transformed, mean, std = transform(train)
test_transformed, _, _ = transform(test, (mean, std))
```

```python
HISTORY = 20
FUTURE_TARGET = 0
trainX, trainY = univariate_data(train_transformed, HISTORY, FUTURE_TARGET)
testX, testY = univariate_data(test_transformed, HISTORY, FUTURE_TARGET)
```

```python
def create_time_steps(length):
    return list(range(-length, 0))


def show_plot(plot_data, delta, title):
    labels = ['History', 'True Future', 'Model Prediction']
    marker = ['.-', 'rx', 'go']
    time_steps = create_time_steps(plot_data[0].shape[0])
    if delta:
        future = delta
    else:
        future = 0

    plt.title(title)
    for i, x in enumerate(plot_data):
        if i:
            plt.plot(future, plot_data[i], marker[i], markersize=10,
               label=labels[i])
        else:
            plt.plot(time_steps, plot_data[i].flatten(), marker[i], label=labels[i])
    plt.legend()
    plt.xlim([time_steps[0], (future+5)*2])
    plt.xlabel('Time-Step')
    return plt
```

```python
from keras.callbacks import Callback
# prepare Keras callback to track training progress in Neptune
class NeptuneMonitor(Callback):
    def __init__(self, neptune_experiment, n_batch):
        super().__init__()
        self.exp = neptune_experiment
        self.n = n_batch
        self.current_epoch = 0

    def on_batch_end(self, batch, logs=None):
        x = (self.current_epoch * self.n) + batch
        self.exp.send_metric(channel_name='batch end loss', x=x, y=logs['loss'])

    def on_epoch_end(self, epoch, logs=None):
        self.exp.send_metric('epoch end loss', logs['loss'])
        self.exp.send_metric('val epoch end loss', logs['val_loss'])
        
        self.current_epoch += 1
```

```python
def nnelu(input):
    """ Computes the Non-Negative Exponential Linear Unit
    """
    return tf.add(tf.constant(1, dtype=tf.float32), tf.nn.elu(input))

tf.keras.utils.get_custom_objects().update({'nnelu': tf.keras.layers.Activation(nnelu)})
```

```python
from tensorflow_probability import distributions as tfd

def slice_parameter_vectors(parameter_vector):
    """ Returns an unpacked list of paramter vectors.
    """
    return [parameter_vector[:,i*components:(i+1)*components] for i in range(no_parameters)]

def gnll_loss(y, parameter_vector):
    """ Computes the mean negative log-likelihood loss of y given the mixture parameters.
    """
    alpha, mu, sigma = slice_parameter_vectors(parameter_vector)  # Unpack parameter vectors
    
    gm = tfd.MixtureSameFamily(
        mixture_distribution=tfd.Categorical(probs=alpha),
        components_distribution=tfd.Normal(
            loc=mu,       
            scale=sigma))
    
    log_likelihood = gm.log_prob(tf.transpose(y))                 # Evaluate log-probability of y
    
    return -tf.reduce_mean(log_likelihood, axis=-1)
```

```python
PARAMS = {'hidden_size': 50,
          'batch_size': BATCH_SIZE,
          'optimizer': "adam",
         'history_steps': HISTORY,
         'epochs': EPOCHS}
```

```python
simple_lstm_model = tf.keras.models.Sequential([
        tf.keras.layers.LSTM(PARAMS["hidden_size"], input_shape=trainX.shape[-2:]),
        tf.keras.layers.Dense(1)])


components = 2
no_parameters = 3
input_1 = tf.keras.layers.Input(shape=trainX.shape[-2:])
lstm = tf.keras.layers.LSTM(PARAMS["hidden_size"])(input_1)
alphas = tf.keras.layers.Dense(components, activation="softmax", name="alphas")(lstm)
mus = tf.keras.layers.Dense(components, name="mus")(lstm)                               
sigmas = tf.keras.layers.Dense(components, activation="nnelu", name="sigmas")(lstm)  
mixture_lstm = tf.keras.layers.Concatenate(name="output")([alphas, mus, sigmas])
mixture_model = tf.keras.models.Model(inputs=[input_1], outputs=[mixture_lstm])
```

```python
model_loss_dict = {"lstm": (simple_lstm_model, "mae"), "mixture_lstm": (mixture_model, gnll_loss) }
```

```python
BATCH_NR
```

```python
choosen_model = "lstm"
model, loss = model_loss_dict[choosen_model]
PARAMS.update({"model":model, "loss":loss})
STEPS_PER_EPOCH = 50
if STEPS_PER_EPOCH and STEPS_PER_EPOCH > BATCH_NR:
    BATCH_NR = STEPS_PER_EPOCH


mon = callbacks.EarlyStopping(monitor='val_loss', min_delta=0, patience=10, verbose=0, mode='auto')
with neptune.create_experiment(params=PARAMS) as exp:
    
    
    EVALUATION_INTERVAL = 300
    EPOCHS = 30
    model.compile(loss=loss, optimizer=PARAMS["optimizer"])
#     history= model.fit(train_univariate, epochs=EPOCHS, validation_data=val_univariate,
#                                    steps_per_epoch=EVALUATION_INTERVAL,
#                                   validation_steps=50, callbacks=[NeptuneMonitor(exp, BATCH_NR)])
    history= model.fit(train_univariate, epochs=EPOCHS, steps_per_epoch=STEPS_PER_EPOCH,
                       validation_data=val_univariate, callbacks=[mon, NeptuneMonitor(exp, BATCH_NR)])
```
